```python
import pandas as pd 
import numpy as np
import random
```

```python
df = pd.read_csv("./Data/credit_customers.csv")
df.head()
```

```python
dfDatosParaComprobar= pd.DataFrame(columns=df.columns)
```

```python
random.seed(922)
random = random.sample(range(0,999),100)
random
```

```python
for i in random:
    if(random.count(i)>1):
        print("Se repite",i)
print("No se repite ningun numero")
```

```python
for i in range (100):
    dfDatosParaComprobar.loc[i]=df.loc[random[i]]
    df.drop(index=random[i], inplace=True)
df.reset_index(drop=True)
```
