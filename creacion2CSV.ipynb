```python
import pandas as pd 
import numpy as np
import random
```

```python
df = pd.read_csv("./Data/Customer-Churn-Records.csv")
df
```

```python
df.drop(columns="RowNumber",inplace=True)
```

```python
df.head()
```

```python
dfDatosParaComprobar= pd.DataFrame(columns=df.columns)
```

```python
random.seed(922)
random = random.sample(range(0,len(df)),500)
random
```

```python
for i in random:
    if(random.count(i)>1):
        print("Se repite",i)
print("No se repite ningun numero")
```

```python
for i in range (500):
    dfDatosParaComprobar.loc[i]=df.loc[random[i]]
    df.drop(index=random[i], inplace=True)
df.reset_index(drop=True)
```

```python
dfDatosParaComprobar.to_csv("./Data/datosParaComprobarModelo.csv",index=False)
```
